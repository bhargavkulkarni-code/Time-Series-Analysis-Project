```python
from statsmodels.tsa.stattools import adfuller
adfuller(petrol['rate'])
```

```python
def adfuller_test(rate):
    result=adfuller(rate)
    label=['adf test statistic','p-value','lags used','number of observations used']
    for t,u in zip(result,label):
        print(u+":"+str(t))
        if result[1]<=0.05:
            print('alternative hypothesis accepted')
        else:
            print('null hypothesis accepted')
```

```python
adfuller_test(petrol['rate'])
```

```python
train=petrol.iloc[:4000,1]
test=petrol.iloc[4000:,1]
```

```python
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
a=ARIMA(petrol['rate'],order=(1,1,0))
b=a.fit()
```

```python
arima_predict=b.predict(len(train),len(train)+len(test)-1,typ='levels')
arima_predict
```

```python
len(arima_predict)
```

```python
q=SARIMAX(petrol['rate'],order=(1,1,0),seasonal_order=(1,1,0,12))
e=q.fit()
```

```python
sarimax_predict=e.predict(len(train),len(train)+len(test)-1,typ='levels')
sarimax_predict
```

# comments:
1. ARIMA is auto regressive integrated moving average it uses in order p,d,q. 
2. p is lagged observations which helps to predict future based on past values.
3. d is differentiation between data points.
4. q is moving average its average of particular years.
5. SARIMAX is seasonal auto regressive integrated moving average  is uses p,d,q,s
6. p,d,q is same meaning what is there in ARIMA.
7. s is addtional component which represents length of seasonal cycle.

```python
# arima plot: test and predicted values
plt.plot(test,color='blue')
plt.plot(arima_predict,color='green')
```

```python
# sarimax plot: test,predicted values
plt.plot(test,color='green')
plt.plot(sarimax_predict,color='purple')
```
